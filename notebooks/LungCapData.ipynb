```python
data.Smoke.replace({'no':0,'yes':1},inplace=True)
data.Gender.replace({'male':1,'female':0},inplace=True)
data.Caesarean.replace({'no':0,'yes':1},inplace=True)                   
```

```python
# Step 1 Independent and Dependent variable creation 

x=data[['LungCap','Age','Height','Smoke','Gender','Caesarean']] # X variable contains all unit.
y=data.Age  ## Y is output.
```

```python
# Step 2 Training and Testing data creation

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
# Step 3 Creating Model

from sklearn.linear_model import LinearRegression
LR = LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
LR.score(x_train,y_train)
```

```python
x_train.shape[1]
```

```python
Rsquare=LR.score(x_train,y_train)
n=x_train.shape[0]
k=x_train.shape[1]
adjrsquare=1-(1-Rsquare)*(n-1)/(n-k-1)
adjrsquare
```

```python
pred_train=LR.predict(x_train)
pred_test=LR.predict(x_test)
err_train=y_train-pred_train
err_train
```
